```python
Employees_Data["degree_clean"]=np.where(Employees_Data["degree"]=="bachelor",1,
                                   np.where(Employees_Data["degree"]=="master",2,
                                            np.where(Employees_Data["degree"]=="phd",3,4)
                                            )
                                  )
```

```python
Employees_Data["education"].value_counts()
```

```python
def education_clean(d):
    if "university" and "institute " in d :
        return 4
    elif "university" and "academy" in d:
        return 3
    elif "university" in d:
        return 2
    else:
        return 1  
    
Employees_Data["education_clean"]=Employees_Data["education"].apply(education_clean)
```

```python
def position_clean(d):
    if d == "dev":
        return 2
    elif d == 'manager':
        return 3
    else:
        return 1
    
Employees_Data["position_clean"]=Employees_Data["position"].apply(position_clean)
Employees_Data["position_clean"].value_counts()
```
